```python
import numpy as np
import pandas as pd

import scipy
from statsmodels.stats.weightstats import *
%pylab inline
```

# BentCigar

```python
datac1 = pd.read_csv('EC_res/Bent_Cau1.csv', sep = ',', header = 0)
datan1 = pd.read_csv('EC_res/Bent_Norm1.csv', sep = ',', header = 0)
datas1 = pd.read_csv('EC_res/Bent_Stud1.csv', sep = ',', header = 0)
```

Diversity

```python
datac1.mean()
```

```python
datan1.mean()
```

```python
datas1.mean()
```

```python
ax = axes()
matplotlib.pyplot.boxplot([datan1.diversity, datac1.diversity, datas1.diversity, ], meanline = True)
matplotlib.pyplot.ylabel('Diversity')
matplotlib.pyplot.title('Bent-Cigar function')
ax.set_xticklabels(['Normal A1\n Mean = 5634', 'Cauchy A2\n Mean = 6267', 'Student A3\n Mean = 5756'])
plt.savefig('Bent_divers.png')
```

```python
ax = axes()
matplotlib.pyplot.boxplot([datan1.score, datac1.score, datas1.score ], meanline = True)
matplotlib.pyplot.ylabel('Score')
matplotlib.pyplot.title('Bent-Cigar function')
ax.set_xticklabels(['Normal A1\n Mean = 9.97', 'Cauchy A2\n Mean = 9.71', 'Student A3\n Mean = 9.94'])
plt.savefig('Bent_score.png')
```

```python
# check normality on qq-plots
stats.probplot(datac1.diversity.values - datan1.diversity.values, dist = "norm", plot = pylab)
pylab.show()
```

```python
stats.probplot(datac1.diversity.values - datas1.diversity.values, dist = "norm", plot = pylab)
pylab.show()
```

```python
stats.probplot(datas1.diversity.values - datan1.diversity.values, dist = "norm", plot = pylab)
pylab.show()
```

```python
stats.ttest_rel(datac1.diversity.values, datan1.diversity.values)
```

```python
stats.ttest_rel(datac1.diversity.values, datas1.diversity.values)
```

```python
stats.ttest_rel(datas1.diversity.values, datan1.diversity.values)
```

```python
print("95%% confidence interval: [%f, %f]" % DescrStatsW(datac1.diversity.values - datan1.diversity.values).tconfint_mean())
print("95%% confidence interval: [%f, %f]" % DescrStatsW(datac1.diversity.values - datas1.diversity.values).tconfint_mean())
print("95%% confidence interval: [%f, %f]" % DescrStatsW(datas1.diversity.values - datan1.diversity.values).tconfint_mean())
```

```python
stats.probplot(datac1.score.values - datan1.score.values, dist = "norm", plot = pylab)
pylab.show()
```
